### 1. Data Imoprt

```python
import pandas
import seaborn
from matplotlib import pyplot
```

```python
# Import the spreadsheet file and assign it to the variable df
df = pandas.read_csv('heart_failure.csv')
```

```python
# Columns in the data file
df.columns
```

### 2. Scatter Plot

```python
seaborn.scatterplot(
    data=df,
    x='age',
    y='serum_creatinine',
)
```

### 3. Line Plot

```python
seaborn.lineplot(
    data=df,
    x='age',
    y='serum_sodium',
)
```

### 4. Histogram

```python
seaborn.displot(
    df,
    x='sex'
).set(
    title='Sex distribution',
    xlabel='Sex',
    ylabel='Frequency'
)
```

### 5. Stacked Histogram

```python
seaborn.displot(
    df,
    x='age',
    bins=range(40, 110, 10),
    hue='diabetes',
    multiple='stack'
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
)
```

### 6. Heatmap

```python
seaborn.displot(
    df,
    x='age',
    y='platelets',
    cbar=True
).set(
    title='Age and Platelets',
    xlabel='Age [years]',
    ylabel='Platelets [platelets/mL]'
)
```

### 7. Box Plot

```python
seaborn.catplot(
    data=df,
    x='smoking',
    y='age',
    
    kind='box'
).set(
    title='Age and Smoking',
    xlabel='Smoking',
    ylabel='Age [years]'
)
```

### 8. Violin Plot

```python
seaborn.catplot(
    data=df,
    x='smoking',
    y='age',
    hue='death',
    kind='violin',
    split=True
).set(
    title='Age and Smoking by Survival Group',
    xlabel='Smoking',
    ylabel='Age [years]'
)
```

### 9. Bar Plot

```python
seaborn.catplot(
    data=df,
    x='smoking',
    hue='death',
    kind='count'
)
```

### 10. Adding a Title

```python
# Create a scatterplot of age vs. serum creatinine and assign it to the variable ax
ax = seaborn.scatterplot(
    data=df,
    x='age',
    y='serum_creatinine',
)
# Add a title to the plot
ax.set_title('My Title')
```

### 11. Saving File

```python
# Save the Age vs. Serum Creatinine scatterplot to a file named output.png
pyplot.savefig('output.png')
# Show the plot
pyplot.show()
```

### 12. Axis Labels

```python
ax = seaborn.lineplot(
    data=df,
    x='age',
    y='serum_creatinine',
)
# Add a title to the plot
ax.set_title('Age vs. Serum Creatinine')
ax.set(xlabel='X Label', ylabel='Y Label')
```

### 13. Changing Color Palette

```python
# Change the color palette of the scatterplot Age vs. Serum Creatinine to 'Blues'
seaborn.scatterplot(
    data=df,
    x='age',
    y='serum_creatinine',
    palette='Blues'
)
```

### 14. Logarithmic Scale

```python
# Create a line plot of age vs. serum sodium and assign it to the variable ax
ax = seaborn.lineplot(
    data=df,
    x='age',
    y='serum_sodium',
)
# Add a title to the plot
ax.set_title('Age vs. Serum Sodium')
ax.set(xlabel='Age [years]', ylabel='Serum Sodium [mEq/L]')
# Add a logarithmic scale to the y-axis
ax.set_yscale('log')
```

### 15. Adding a Regression Line

```python
seaborn.regplot(
    data=df,
    x='age',
    y='serum_creatinine',
    line_kws={'color': 'red', 'label': 'Regression Line'},
    scatter_kws={'color': 'pink', 'label': 'Data Points'}
)
```
