## Analyzing 2018 Total Spending Data

```python
# Convert the total_spending2018_column to a numpy array and assign it to the variable total 2018 spending
total_2018_spending = total_spending2018_column.to_numpy()
```

```python
# Print the type of total 2018 spending
type(total_2018_spending)
```

#### The Average Medicaid Spending For All Drugs in 2018 

```python
# Calculate the average spending in total_2018_spending
total_spending2018_column.mean()
print(total_spending2018_column.mean())
```

#### The Minimum Medicaid Spending For Any Drug in 2018

```python
# Calculate the minimum spending in total_2018_spending_column
total_spending2018_column.min()
print(total_spending2018_column.min())
```

```python
# Identify the Brnd_Name column in df with the minimum spending in total_2018_spending_column
df[df['Tot_Spndng_2018'] == total_spending2018_column.min()]['Brnd_Name']
```

##### Analysis: Adhesive tape, bandages, bandages sheer, clear bandages, plastic bandages, and simvastatin are the brand names that have the mimum spending recorded in the 2018 total spending column.

#### The Maximum Medicaid Spending For All Drugs in 2018

```python
# Calculate the maximum spending in total_2018_spending_column
total_spending2018_column.max()
print(total_spending2018_column.max())
```

```python
# Identify the Brnd_Name column in df with the maximum spending in total_2018_spending_column
df[df['Tot_Spndng_2018'] == total_spending2018_column.max()]['Brnd_Name']
```

##### Analysis: The Humira Pen is the brand name with the highest spending of total 2018 spending.
